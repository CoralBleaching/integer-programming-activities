Agora tentando com um grafo mais substancial. Não é útil fazer um plot da árvore de decisão do BAB neste caso.

```python
gg5 = vc.generate_graph(np.random.default_rng(2023), 50, 0.3)
sol5, tree5 = vc.solve_bab_vertex_cover(gg5)
solver5 = vc.generate_vertex_cover_problem(gg5, solver_version='SCIP')
if solver5 is None: raise RuntimeError("solver5 não foi instanciado.")
solver5.Solve()
sol5scip = tuple(var.solution_value() for var in solver5.variables())
```

```python
print(tabulate([
    ['`solve_bab_vertex_cover`', sum(sol5)],
    ['SCIP', sum(sol5scip)]
    ],
    headers=['Rotina', 'Valor objetivo']))
```

```python
vc.plot_solved_graph(sol5, gg5)
```

```python
vc.plot_solved_graph(sol5scip, gg5)
```
